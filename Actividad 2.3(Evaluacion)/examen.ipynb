```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

**CARGA DE ARCHIVO .CSV**

```python
#convertir en dataframe 
df=pd.read_csv("Victoria.csv")
df
```

```python
#mostramos información del dataframe 
df.info()
```

**a. Registros de 5 "host_name diferentes**

```python
#los registros de 5 "host_name" diferentes
#aplicamos filtro por objeto y filtro or para buscar en nuestro archivo todas las filas que contengan los 5 nombres diferentes
f1=df[(df["host_name"] == "Martin") | (df["host_name"] == "Ryan") | (df["host_name"] == "Ann") | (df["host_name"] == "Gurpaul Singh") | (df["host_name"] == "Line Marie")]
f1
```

b. Registros de los host que se hayan unido a Airbnb "antes del año 2019" 

```python
#los registros de los host que se hayan unido a Airbnb "antes del año 2019"
#primero convertimos la columna host_since a tipo datatime y posteriormente aplicamos un filtro por comparación para extraer solo los datos de los host que se unieron antes del 2019
df["host_since"] = pd.to_datetime(df["host_since"], format="%d/%m/%Y", errors="coerce", dayfirst=True)
f2=df[df["host_since"] < "01/01/2019"]
f2
```

c. Registros de los host que responden a más tardar en un día 

```python
#los registros de los hosts que responden a más tardar en un día 
#en este filtro trabajamos con la variable host_response_time para los que responden a más tardar en un día 
#f3=df[(df["host_response_time"] == "within a day")] #tarda un día 
f3=df[(df["host_response_time"] =="whithin a day") | (df["host_response_time"] == "within an hour") | (df["host_response_time"] == "within a few hours")]
f3
```

d. registros que tengan al menos 5 accommodates 

```python
#los registros que tengan "al menos 5 accommodates"
#aplicamos un filtro por comparación a la variable accommodates para solo extraer los que sean >=5
f4=df[(df["accommodates"] >=5)]
f4
```

e. registros de los tipos de cuarto: "Entire home" y "Private room"

```python
#los registros de los tipos de cuarto: "Entire home" y "Private room"
#en este paso aplicaremos filtro a la variable "room_type" para filtrar los datos que sean "Entire home" y "Private room" exclusivamente 
f5=df[(df["room_type"] =="Entire home/apt") | (df["room_type"] == "Private room")]
f5
```

f. Registros que a lo mucho cuentan con 4 camas 

```python
#los registros que "a lo mucho cuentan con 4 camas"
#aplicamos a la variable beds un filtro de compración que sea <=4 para encontrar solo los registros que cuenten con 4 camas o menos 
f6=df[(df["beds"] <= 4)]
f6
```

g. Registros que tengan 1 baño o sean super_host pero con la condición de que todos sean instant_bookable

```python
#los registros que tengan 1 baño o sean super_host pero con la condición de que todos sean instant_bookable
#aplicamos un filtro combinado de and y or, or para los regisros que tengan 1 baño o sean super_host, and para los registros que sean instant_bookable 
f7 = df[((df["bathrooms"] == 1) | (df["host_is_superhost"] == "t")) & (df["instant_bookable"] == "t")]
f7
```

h. registros cuyos review_scores_cleanliness sea menor de 3 y review_scores_rating sea mayor de 3

```python
#los registros cuyo review_scores_cleanliness sea menor de 3 y review_scores_rating sea mayor de 3
#en este filtro aplicaremos un filtro and para obtener los registros que cumplan ambas condiciones 
f8=df[(df["review_scores_cleanliness"] < 3) & (df["review_scores_rating"] > 3)]
f8
```
