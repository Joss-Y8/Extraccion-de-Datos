```python
#librerias requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as ns
```

**CARGA DE ARCHIVO XLSX**

```python
#carga desde un archivo .xlsx sin índice
df= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#verificamos información del DataFrame
df.info()
```

**FILTROS BÁSICOS PARA VARIABLES CUALITATIVAS**

```python
#filtro por objeto, en este filtro se filtran los datos que se encuentren para el cliente ("NOMBRE_CLIENTE") registrado como FIDEICOMISO F/1596
f1=df[(df["NOMBRE_CLIENTE"] == "FIDEICOMISO F/1596")]
f1
```

```python
#filtro por filas, en este filtro se extraen los datos que contienen las filas 2, 5 y 6 dentro de nuestra base de datos con toda la información que contienen las varibles
#f2=df.iloc[2:8 , :]
f2=df.iloc[ [2,5,6] , :]
f2
```

```python
#filtro por columnas, aquí se trabaja con las columnas 5, 6, 7 y 8 esta indicación solo extrae las columnas con las varibles CVE_ART, DESCR, PRECIO_UNITARIO Y COSTO_UNITARIO
f3=df.iloc[: , 5:9]
#f3=df.iloc[: , [2 , 5]]
f3
```

```python
#Carga de archivo .xlsx con indice 
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=3)
df1
```

```python
#Paso 2: se aplica filtro de filas vs columnas.
#Para aplicar este filtro primero se debe cargar el archvio .xlsx  con indice de la columna donde se encuentra
# la variable que quieres extraer en este caso NOMBRE_CLIENTE, se mostraran las filas que contienen los datos seleccionados en ese indice
f4=df1.loc[["ELICAMEX", "VALEO KAPEC"], ["MARGEN_TOTAL_CALCULADO"]]
f4
```

```python
#filtro por cabecera, mostrará las primeras 10 filas extraidas de la base de datos (cuando el parentesis esté vacío muestra las primeras 5)
f5=df.head(10)
f5
```

```python
#filtro por cola, mostrará las últimas 10 filas extraidas de la base de datos (cuando el parentesis esté vacío muestra las últimas 5)
f6=df.tail(10)
f6
```

**FILTROS BÁSICOS PARA VARIABLES CUANTITATIVAS**

```python
#filtro por comparación, en este flitro se selecciona una variable de la que se requiera extraer datos con uns criterio especificado. 
f7=df[df["COSTO_UNITARIO"] > 1500]
f7
```

**FILTROS BÁSICOS CON CONECTORES LÓGICOS**

```python
#filtro y, se trabaja con el conector lógico & para extraer datos específicos en FECHAS_DOC con MARGEN_TOTAL_CALCULADO que contiene los datos menores a 6000
f8=df[(df["MARGEN_TOTAL_CALCULADO"] < 6000) & (df["FECHA_DOC"] == "2022-12-23")]
f8
```

```python
#filtro o, a diferencia del conector &, | muestra los datos que en MARGEN_TOTAL_CALCULADO son <6000 o los datos de FECHA_DOC que se han específicado 
f9=df[(df["MARGEN_TOTAL_CALCULADO"] < 6000) | (df["FECHA_DOC"] == "2022-12-23")]
f9
```

```python
#filtro not, en este filtro trabajo con la varibale NOMBRE_VENDEDOR para extraer los datos que no hayan sido obtenidos del vendedor especificado. 
f10= df[~(df["NOMBRE_VENDEDOR"] == "LETICIA RAMIREZ HERNANDEZ")]
f10
```
